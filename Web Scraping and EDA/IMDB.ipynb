```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
from selenium import webdriver
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
movie_name = []
year = []
runtime = []
certificate = []
rating = []
voteCount = []
metaScore = []
director = []

for i in range(1,5):
    driver.get(f"https://www.imdb.com/list/ls540232479/?page={i}")
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(3)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height


    html = driver.page_source
    soup = BeautifulSoup(html)

    for i in soup.find_all("div", class_ = "sc-2bfd043a-3 jpWwpQ dli-parent"):
        text = i.text

        title = i.find("h3", class_ = "ipc-title__text")

        if title:
            nm = title.text

        r1 = re.findall(" ([0-9A-Z].+)", nm)
        if r1:
            movie_name.append(r1[0])
        else:
            movie_name.append(np.nan)


        details = i.find('div', class_ = "sc-300a8231-6 dBUjvq dli-title-metadata")

        if details:
            dt = details.text

            r2 = re.findall("[0-9]{4}", dt)
            if r2:
                year.append(r2[0])
                
            else:
                year.append(np.nan)    

            r3 = re.findall("\dh \d+m|\dh|..m", dt)
            if r3:
                runtime.append(r3[0])
            else:
                runtime.append(np.nan)   

            r4 = re.findall("m(.*)", dt)
            if r4:
                certificate.append(r4[0])
            else:
                certificate.append(np.nan)

        rat_vot = i.find("span", class_ = "sc-300a8231-1 kWSYcu")

        if rat_vot:
            rv = rat_vot.text

            r5 = re.findall("(\d\.\d)",rv)
            if r5:
                rating.append(r5[0])
            else:
                rating.append(np.nan)

            r6 = re.findall("\((.+?)\)",rv)
            if r6:
                voteCount.append(r6[0])
            else:
                voteCount.append(np.nan)

        meta = i.find("span", class_ = "sc-300a8231-9 fNNdeo")

        if meta:
            ms = meta.text

            r7 = re.findall("[0-9]+", ms)
            if r7:
                metaScore.append(r7[0])
        else:
            metaScore.append(np.nan)

        dire = i.find("span", class_ = "sc-2bfd043a-5 fIttnm")
        if dire:
            director.append(dire.text)
        else:
            director.append(np.nan)
                     
```
